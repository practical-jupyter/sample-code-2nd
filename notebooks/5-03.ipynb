## 5.3: グラフの軸と目盛りの設定

```python
# リスト 5.3.1　軸の書式を設定せずに描画（デフォルト）

import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
x1 = np.arange(0, 10, 0.2)
x2 = np.arange(0, 20, 0.2)
x3 = np.arange(0, 30, 0.2)
x4 = np.arange(0, 40, 0.2)
y1 = np.sin(x1)
y2 = np.sin(x2) * 1.5
y3 = np.cos(x3) * 2
y4 = np.cos(x4)
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(x1, y1)
axes[0, 1].plot(x2, y2)
axes[1, 0].plot(x3, y3)
axes[1, 1].plot(x4, y4)
```

```python
# リスト 5.3.2　sharex, sharey 引数を設定して描画
fig, axes = plt.subplots(2, 2, sharex=True, sharey="col")

axes[0, 0].plot(x1, y1)
axes[0, 1].plot(x2, y2)
axes[1, 0].plot(x3, y3)
axes[1, 1].plot(x4, y4)
```

```python
# リスト 5.3.3　軸の範囲設定

fig, axes = plt.subplots(2, 2)

# 折れ線グラフの描画
axes[0, 0].plot(x1, y1)
axes[0, 1].plot(x2, y2)
axes[1, 0].plot(x3, y3)
axes[1, 1].plot(x4, y4)

# X軸の範囲の設定
axes[0, 0].set_xlim(0, 10)
axes[0, 1].set_xlim(10, 20)
axes[1, 0].set_xlim(20, 10)
axes[1, 1].set_xlim(10, 0)

# Y軸の範囲の設定
axes[0, 0].set_ylim(0, 1)
axes[0, 1].set_ylim(0, -1)
axes[1, 0].set_ylim(-1, 0)
axes[1, 1].set_ylim(1, 0)

# サブプロット間スペースの調整
fig.subplots_adjust(hspace=0.3, wspace=0.3)
```

```python
# リスト 5.3.4　対数目盛りの設定

a = np.linspace(0, 100, 100)
b = np.linspace(0, 1000000, 100)

fig, axes = plt.subplots(1, 2)

axes[0].plot(a, b)
axes[1].plot(b, a)

# 軸を対数に設定
axes[0].set_yscale("log")
axes[1].set_xscale("log")

axes[0].set_title("Y軸が対数軸")
axes[1].set_title("X軸が対数軸");
```

```python
# リスト 5.3.5　複数軸を設定

p = np.linspace(-10, 10, 20)
q = np.linspace(0, 10, 20)
r = np.linspace(-10, 10, 50)
s = [(np.tanh(ss / 2) + 1) / 2 for ss in r]
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# 左図の描画
# 左軸を持つデータを描画
axes[0].scatter(p, q, color="blue", label="axes[0]")

# 右軸を持つデータを描画
ax2 = axes[0].twinx()
ax2.plot(r, s, color="blue", label="ax2")

# 右図の描画
# 下軸を持つデータを描画
axes[1].scatter(q, p, color="red", label="axes[1]")

# 上軸を持つデータを描画
ax3 = axes[1].twiny()
ax3.plot(s, r, color="red", label="ax3")

# サブプロットタイトルの設定
axes[0].set_title("2つのY軸")
axes[1].set_title("2つのX軸")

# 凡例の設定
fig.legend();
```

```python
# リスト 5.3.6　軸ラベルの設定

fig, ax1 = plt.subplots(1, 1)
ax1.plot(p, q, color="blue", label="ax1")
ax2 = ax1.twinx()
ax2.plot(r, s, color="red", label="ax2")
ax1.set_ylabel("ax1", labelpad=-5, fontsize=20, color="blue")
ax2.set_ylabel("ax2", {"fontsize": 20, "color": "red"})
fig.legend(bbox_to_anchor=(0.85, 0.25))
```

```python
# リスト 5.3.7　主目盛りと主目盛りラベルを設定

fig, ax1 = plt.subplots(1, 1)
ax1.plot(p, q, color="blue", label="ax1")
ax2 = ax1.twinx()
ax2.plot(r, s, color="red", label="ax2")

# 主目盛りの設定
ax1.set_xticks([-10, 0, 10])
ax1.set_yticks(np.linspace(0, 10, 3))
ax2.set_yticks(np.linspace(0, 1, 3))

# 主目盛りラベルの設定
ax1.set_yticklabels(["zero", "five", "ten"])
ax1.set_xticklabels([])
fig.legend(bbox_to_anchor=(0.85, 0.25))
```

```python
# リスト 5.3.8　目盛りの書式設定①

fig, ax1 = plt.subplots(1, 1)
ax1.plot(p, q, color="blue", label="ax1")
ax2 = ax1.twinx()
ax2.plot(r, s, color="red", label="ax2")

# X軸の主目盛りを設定
ax1.set_xticks([-10, 0, 10])

# 補助目盛りの描画
ax1.minorticks_on()

# 目盛り線の描画
ax1.grid()

# Y軸（左軸）の主目盛りを設定
ax1.tick_params(
    axis="y",
    which="major",
    direction="inout",
    length=10,
    width=3,
    color="blue",
    labelcolor="blue",
    labelrotation=90,
)

# Y軸（右軸）の主目盛りを設定
ax2.tick_params(
    axis="y",
    which="major",
    direction="inout",
    length=10,
    width=3,
    color="red",
    labelcolor="red",
    labelrotation=-90,
)

# 左軸の補助目盛りを設定
ax1.tick_params(
    axis="y",
    which="minor",
    direction="inout",
    length=10,
    color="blue",
    bottom="on",
    grid_color="blue",
    grid_linewidth=1,
    grid_linestyle="-",
)

# X軸の主目盛りを設定
ax1.tick_params(
    axis="x",
    which="major",
    direction="inout",
    length=10,
    width=3,
    right="on",
    grid_linewidth=0,
)

# X軸の補助目盛りを設定
ax1.tick_params(axis="x", which="minor", direction="inout", length=5, right="on")
fig.legend(bbox_to_anchor=(0.8, 0.3))
```

```python
# リスト 5.3.9　目盛りの書式設定②

fig, ax1 = plt.subplots(1, 1)
ax1.plot(p, q, color="blue", label="ax1")
ax2 = ax1.twinx()
ax2.plot(r, s, color="red", label="ax2")

# 補助目盛りの描画
ax1.minorticks_on()
ax2.minorticks_on()

# X軸の主目盛りを設定
ax1.set_xticks([-10, 0, 10])

# 目盛り線の描画
ax1.grid(which="major", axis="x", color="black", linestyle="-", linewidth=1)
ax1.grid(which="minor", axis="x", color="grey", linestyle="--", linewidth=1)
ax1.grid(which="major", axis="y", color="grey", linestyle="--", linewidth=1)
```

```python
# リスト 5.3.10　軸線の設定

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# 折れ線グラフを描画
s_array = np.array(s) - 0.5
axes[0].plot(r, s_array, color="blue")
axes[1].plot(r, s_array, color="green")

# 軸線の色の指定
axes[0].spines["left"].set_color("blue")
axes[0].spines["bottom"].set_color("blue")
axes[0].spines["right"].set_color(None)
axes[0].spines["top"].set_color(None)
axes[1].spines["left"].set_color("green")
axes[1].spines["bottom"].set_color("green")
axes[1].spines["right"].set_color(None)
axes[1].spines["top"].set_color(None)

# 軸線の描画位置の指定
axes[0].spines["left"].set_position(("outward", 5))
axes[0].spines["bottom"].set_position("center")
axes[1].spines["left"].set_position(("data", -5))
axes[1].spines["bottom"].set_position("zero")

# Y軸の描画範囲を指定（左図）
axes[0].set_ylim([-0.6, 0.6])

# 軸線の描画範囲を指定（左図）
axes[0].spines["left"].set_bounds(-0.6, 0.6)
axes[0].spines["bottom"].set_bounds(-10, 10)
```
